```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv("/content/Dataset - Dataset (1).csv")
print("Initial Dataset Preview:")
data.head()


cols_to_remove = [col for col in ["id", "Unnamed: 32"] if col in data.columns]
data = data.drop(columns=cols_to_remove)


data["diagnosis"] = data["diagnosis"].map({"M": 1, "B": 0})

data["diagnosis"].value_counts()
data.head()


normalizer = MinMaxScaler()

features = data.drop("diagnosis", axis=1)
target = data["diagnosis"]

scaled_features = normalizer.fit_transform(features)


X_train_set, X_test_set, y_train_set, y_test_set = train_test_split(
    scaled_features, target, test_size=0.20, random_state=42
)


kmeans_model = KMeans(n_clusters=2, random_state=42)
cluster_labels = kmeans_model.fit_predict(scaled_features)

data["cluster_group"] = cluster_labels

pd.crosstab(data["cluster_group"], data["diagnosis"])


knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_set, y_train_set)

y_predicted = knn_model.predict(X_test_set)


accuracy = accuracy_score(y_test_set, y_predicted)
precision = precision_score(y_test_set, y_predicted)
recall = recall_score(y_test_set, y_predicted)
f1_val = f1_score(y_test_set, y_predicted)

print("Accuracy  :", accuracy)
print("Precision :", precision)
print("Recall    :", recall)
print("F1-score  :", f1_val)
```
